**T Test Examples**
1. A coaching institute claims that the average score of its students in Statistics is 75 marks. A random sample of 12
students shows a mean score of 71 marks with a standard deviation of 8 marks.
At 5% significance level, test whether the claim is correct (Two Tailed)

#Given
1. Sample count= 12 students
2. Sample Mean = 71 Marks
3. Population Mean= 75 Marks
4. Standard Deviation : 8 Marks
5. Significance Level : 5%

# Analysis
1. Null Hypothesis H0= mean is equal to 75
2. Alternate Hypothesis H1= Mean is not equal to 75

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
def t_test_two_tailed(n,x,u,s,alpha):
  df=n-1
  t=(x-u)/(s/np.sqrt(n))
  print("t-statistic=",t)
  t_critical=stats.t.ppf(1-alpha/2,df)
  t_critical=abs(t_critical)
  print("t-critical=",t_critical)
  if t>t_critical:
    print("Reject Null Hypothesis")
  else:
    print("Accept Null Hypothesis")

def t_test_left_tailed(n,x,u,s,alpha):
  df=n-1
  t=(x-u)/(s/np.sqrt(n))
  print("t-statistic=",t)
  t_critical=stats.t.ppf(alpha,df)
  print("t-critical=",t_critical)
  if t>t_critical:
    print("Reject Null Hypothesis")
  else:
    print("Accept Null Hypothesis")
def t_test_right_tailed(n,x,u,s,alpha):
  df=n-1
  t=(x-u)/(s/np.sqrt(n))
  print("t-statistic=",t)
  t_critical=stats.t.ppf(1-alpha,df)
  print("t-critical=",t_critical)
  if t>t_critical:
    print("Reject Null Hypothesis")

  else:
    print("Accept Null Hypothesis")
```

```python
t_test_two_tailed(12,71,75,8,0.05)
```

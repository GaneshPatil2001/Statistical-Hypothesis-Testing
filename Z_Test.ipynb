##Two-Tailed Z-Test Problems (Hypothesis Testing)
1.The average height of students (500) in a college is believed to be 165 cm.
A random sample of 64 students has a mean height of 168 cm.
The population standard deviation is 12 cm.
At 5% level of significance, test whether the average height differs from 165 cm.

#Given
1. Total Students = 500
2. Sample Students = 64
3. Population Mean = 165
4. Sample Mean = 168
5. Population Standard deviation = 12
6. Level Of Siginficanc = 5%



```python
import numpy as np
from scipy.stats import norm

# Given values from the problem statement
population_mean = 165
sample_mean = 168
population_std = 12
sample_size = 64
alpha = 0.05


standard_error = population_std / np.sqrt(sample_size)
z_score = (sample_mean - population_mean) / standard_error

p_value = 2 * norm.sf(abs(z_score)) # sf is the survival function (1 - CDF)

print(f"Z-Statistic: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
  print(f"Since P-Value ({p_value:.4f}) < Alpha ({alpha}), we Reject the null hypothesis.")
  print("Conclusion: The average height differs from 165 cm.")
else:
  print(f"Since P-Value ({p_value:.4f}) >= Alpha ({alpha}), we Fail to reject the null hypothesis.")
  print("Conclusion: The average height does not differ significantly from 165 cm.")
```
